```python
import pandas as pd
import numpy as np
```

```python
# pivot table 
# in pivot table aggregation func. use just like mean, max, mode and so on.
```

```python
data = {
    'Date' : pd.date_range('2023-01-01',periods = 20),
    'Product' : ['A', 'B', 'C', 'D'] * 5,
    'Region' : ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales' : np.random.randint(100, 1000, 20),
    'Units' : np.random.randint(10, 1000, 20),
    'Rep' :['John', 'Marry', 'Larry', 'Elon', 'John', 'Elon', 'Marry', 'Larry', 'John', 'Elon',
           'Larry', 'Marry', 'Elon', 'John', 'Marry', 'Larry', 'Elon', 'John', 'Marry', 'Larry']
}
df =  pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values = "Sales", index = 'Region', columns="Product", aggfunc = 'median')
```

```python
pivot2 = pd.pivot_table(df,values = ["Sales","Units"], index = 'Region', columns="Product", aggfunc = 'median')
pivot2
```
